```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime 
basic_folder = '/Users/yangj/'
file_name = basic_folder + '국토안전관리원_건설안전사고 발생현황1.csv'
df = pd.read_csv(file_name,encoding='euc-kr')
df
```

```python
df['사고신고사유1']=df.사고신고사유.str.split(',').str[1]
df['사고신고사유2']=df.사고신고사유.str.split(',').str[0]
df['사고날짜']=df.사고일시.str.split(' ').str[0]
df['사고시간']=df.사고일시.str.split(' ').str[1]
new_df=df[['사고신고사유1','사고신고사유2','사고날짜','사고시간','공정율','날씨','공사종류','공사비','공종(소분류)','작업자수','사망자','부상자']]
df_month=new_df.copy()
df_month['datetime']=pd.to_datetime(new_df['사고날짜'])
df_month['month']=df_month['datetime'].dt.month
condition =[
    (df_month['month']>=2) & (df_month['month']==12),
    (df_month['month']>=3) & (df_month['month']<=5),
    (df_month['month']>=6) & (df_month['month']<=8),
    (df_month['month']>=9) & (df_month['month']<=11)
]
choice=['겨울','봄','여름','가을']
df_month['계절']=np.select(condition,choice,default='nan')
df_time=df_month.copy()
df_time['time']=pd.to_datetime(df_month['사고시간'])
df_time['hour']=df_time['time'].dt.hour
df_time.drop(['month','time'],axis=1,inplace=True)
df_time
```

```python
df_die=df_time[['datetime','계절','사고시간','hour','공정율','날씨','공사종류','공종(소분류)','공사비','작업자수','사망자']]
die=df_die['사망자']>0
only_die=df_die[die]
only_die
```

```python
n_die = only_die.copy()
n_die['최소공사비']=df.공사비.str.split('~').str[0]
n_die
```

```python
n_die.to_csv("사망자발생사고_ver1202.csv")
```

```python
df_injury=df_time[['datetime','계절','사고시간','hour','공정율','날씨','공사종류','공종(소분류)','공사비','작업자수','부상자']]
injury=df_injury['부상자']>0
only_injury=df_injury[injury]
n_injury = only_injury.copy()
n_injury['최소공사비']=df.공사비.str.split('~').str[0]
n_injury
```

```python
n_injury.to_csv("부상자발생사고_ver1202.csv")
```
